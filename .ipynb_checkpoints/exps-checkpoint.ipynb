```python
import pandas as pd
```

```python
df=pd.read_csv("data/train_data.csv")
df["target_labels_positions"]=df["target_labels_positions"].map(eval)
df["tokens"]=df["processed_text"].str.split(" ")
df["count_of_tokens"]=df["tokens"].map(len)
df["labels"]=df["count_of_tokens"].map(lambda x: x*['O'])
for i in df.index:
    for key in df.loc[i,"target_labels_positions"].keys():
        for pos in df.loc[i,"target_labels_positions"][key]:
            df.loc[i,"labels"][pos]=key
df
```

```python
vocabular = set()
for row in df.index:
    vocabular = vocabular.union(set(df.loc[row,"tokens"]))
len(vocabular)                   
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder_voc=LabelEncoder()
encoder_voc.fit(list(vocabular))
encoder_voc.classes_
```

```python
df['tokens_codes']=df['tokens'].map(encoder_voc.transform)
df
```

```python
labels = set()
for row in df.index:
    labels = labels.union(set(df.loc[row,"labels"]))
len(labels) 
```

```python
labels
```

```python
encoder_label=LabelEncoder()
encoder_label.fit(list(labels))
encoder_label.classes_
```

```python
df["labels_codes"]=df["labels"].map(encoder_label.transform)
df
```
